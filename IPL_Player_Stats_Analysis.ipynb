# IPL Cricket Player Stats Analysis
This project analyzes IPL player performance stats using pandas and visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('ipl_player_stats.csv')
df.head()
```

```python
# Top 5 run scorers
top_scorers = df.sort_values(by='Runs', ascending=False).head()
print(top_scorers[['Player', 'Runs']])
```

```python
# Plot Runs
plt.figure(figsize=(8,5))
sns.barplot(x='Player', y='Runs', data=top_scorers)
plt.title('Top 5 Run Scorers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
